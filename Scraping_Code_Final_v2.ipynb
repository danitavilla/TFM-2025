```python
# Importamos librerias necesarias
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# Funcion para extraer los detalles de cada publicacion
def publication_object(car_publication):
    try:
        # Extraer marca y modelo
        marca_modelo = car_publication.find(class_="card-product_cardProduct__title__RR0CK").text.strip()
        marca, modelo = marca_modelo.split(" • ") if " • " in marca_modelo else (marca_modelo, None)
        # Extraer año, kilometraje, version y transmision
        subtitulo = car_publication.find(class_="card-product_cardProduct__subtitle__hbN2a").text.strip()
        subtitulo_parts = subtitulo.split(" • ")
        año = subtitulo_parts[0] if len(subtitulo_parts) > 0 else None
        km = subtitulo_parts[1].replace(",", "") if len(subtitulo_parts) > 1 else None
        version = subtitulo_parts[2] if len(subtitulo_parts) > 2 else None
        transmision = subtitulo_parts[3] if len(subtitulo_parts) > 3 else None
        
        kilometraje = km.split(" ")
        kilometraje = km[0]
                
        # Extraer precio
        precio = car_publication.find(class_="uki-amount").text.strip().replace("$", "").replace(",", "")
       
        # Extraer ubicacion del vehiculo
        ubicacion = car_publication.find(class_="card-product_cardProduct__footer__MQKa_").text.strip()

        # Asignar nombre a las columnas
        return {
            "marca": marca,
            "modelo": modelo,
            "año": año,
            "kilometraje": kilometraje,
            "version": version,
            "transmision": transmision,
            "precio (MXN)": precio,
            "ubicacion": ubicacion
        }
    except Exception as e:
        print(f"Error extracting data: {e}")
        return {}
```

```python
# Funcion para hacer scrape y guardar las publicaciones de una pagina
def scrape_page(url):
    response = requests.get(url, headers={"User-Agent": "Mozilla/5.0"})
    
    if response.status_code != 200:
        print(f"Failed to retrieve page {url}")
        return []

    soup = BeautifulSoup(response.text, "html.parser")
    all_publications = soup.find_all("a", class_="card-product_cardProduct__3qZCJ")

   # return [publication_object(car) for car in all_publications]

    # Filtrar solo publicaciones con datos válidos
    return [car for car in (publication_object(car) for car in all_publications) if car]
```

```python
# Iterar a traves de las diferentes paginas para hacer el scraping (paginas del 0 al 150)
base_url = "https://www.kavak.com/mx/seminuevos?page="
all_cars = []

for page in range(0, 150):
    print(f"Scraping page {page}...")
    all_cars.extend(scrape_page(base_url + str(page)))
```

```python
if all_cars:
    df = pd.DataFrame(all_cars)
    df.to_csv("cars_publications.csv", mode='a', index=False, encoding="utf-8-sig")
    print("Scraping completed. Data saved.")
else:
    print("No valid data extracted.")
```
